```python
import seaborn as sns
sns.heatmap(tdata.isnull())
```

```python
ffilldata = tdata.fillna(method="ffill")
```

```python
sns.heatmap(ffilldata.isnull())
```

```python
import matplotlib.pyplot as plt
```

```python
bc = ffilldata.groupby('Embarked')['Embarked'].count()
```

```python
bc
```

```python
plt.figure()
plt.pie(bc.values)
```

```python
plt.pie(bc.values, labels=['C TOWN','Q TOWN','S TOWN'])
```

```python
plt.title("Boarding Station")
plt.pie(bc.values, labels=['C TOWN','Q TOWN','S TOWN'])
```

```python
plt.title("Boarding Station")
exp = [0.1,0,0]
plt.pie(bc.values, labels=['C TOWN','Q TOWN','S TOWN'], explode=exp, shadow=True)
```

```python
plt.figure(figsize=(10,10))
plt.title("Boarding Station")
exp = [0.1,0,0]
plt.pie(bc.values, labels=['C TOWN','Q TOWN','S TOWN'], explode=exp, shadow=True)
```

```python
plt.title("Boarding Station")
exp = [0.1,0,0]
plt.pie(bc.values, labels=['C TOWN','Q TOWN','S TOWN'], explode=exp, shadow=True)
```

```python
plt.title("Boarding Station")
exp = [0.1,0,0]
plt.pie(bc.values, labels=['C TOWN','Q TOWN','S TOWN'], explode=exp, shadow=True, startangle=90)
```

```python
plt.title("Boarding Station")
exp = [0.1,0,0]
plt.pie(bc.values, labels=['C TOWN','Q TOWN','S TOWN'], explode=exp, shadow=True, colors=["Violet","Red","Black"])
```
